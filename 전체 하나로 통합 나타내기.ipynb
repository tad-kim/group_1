```python
import numpy as np
import csv
import re

with open ("C:\\Users\\zkdlt\\Desktop\\p\\IMDB top 1000.csv") as f:
    reader = csv.reader(f)
    data_list = list(reader)
# numpy 배열로 변환
content = np.array(data_list[1:])
header = data_list[0]

# 3개의 컬럼만 추출
title_idx = header.index('Title')
genre_idx = header.index('Genre')
rate_idx = header.index('Rate')
data_cul = content[:,[title_idx,genre_idx,rate_idx]]

# Title 앞에 숫자 제거
data_cul[0] = [re.sub(r'\d+\.\s*', '', title) for title in data_cul[0]]

# 연도 추출하여 Released_Year 컬럼 추가
Released_Year = []
for i in data_cul[:,0]:
    if '(' in i and ')' in i and len(i) >= 6:
        Released_Year.append(i[-5:-1])
    else:
        Released_Year.append('')
year_np = np.array(Released_Year).reshape(-1,1)
#결측값 제거
cler = []
for j in data_cul:
    if '' not in j:
        cler.append(j)
data_cul = np.array(cler)
df_total = np.hstack((data_cul, year_np))
# 데이터 출력 및 데이터 크기
print(df_total[:5])
print(df_total.shape)

# Step2
# 전체 영화의 개수 출력 및 평점 
print("총 영화 개수:", len(df_total))
avg1 = np.mean(df_total[:,2].astype(float))
print(f"평점 평균: {avg1:.1f}")
max1 = np.max(df_total[:,2].astype(float))
print(f"최고 평점: {max1}")
min1 = np.min(df_total[:,2].astype(float))
print(f"최저 평점: {min1}")

# Step3
# 데이터에서 최고 평점(최대값 찾기)
# 해당 평점을 받은 영화 목록을 출력하기
avg2 = df_total[:,2].astype(float)
best_movie = np.where(avg2 == max1)[0]
print("★ 최고 평점 영화 목록:")
for idx in best_movie:
    print(f"{df_total[idx,0]}-평점:{df_total[idx,2]}")

# Step4
# 장르별 평균 평점 계산
genre_list = {}
for i in df_total:
    gen = i[1].split(',')
    for j in gen:
        j = j.strip()
        genre_list.setdefault(j,[]).append(float(i[2]))
# 제목 평점 튜플로 만들기
genre_avg = []
for gen1 in genre_list:
    avg3 = sum(genre_list[gen1])/len(genre_list[gen1])
    genre_avg.append((gen1, avg3))
# 평점 높은순 정렬
genre_bast = sorted(genre_avg, key=lambda x: x[1], reverse=True)
print("장르별 평균 평점:")
for gen2, avg4 in genre_bast:
    print(f"{gen2}:{avg4:.1f}")
```
